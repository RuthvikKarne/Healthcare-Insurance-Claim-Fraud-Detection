```python
import pandas as pd
import numpy as np
```

```python
data=pd.read_excel(r'D:\1.Tekworks\Project\Healthcare Insurance Claim\Backend\1.Dataset\Clean data\clean_data.xlsx')
```

```python
data.head()
```

```python
boolean_columns = [
    'Claim_Submitted_Late',
    'Is_Fraudulent'
]
data['Claim_Submitted_Late'] = data['Claim_Submitted_Late'].astype(int)

data['Is_Fraudulent'] = data['Is_Fraudulent'].astype(int)
```

```python
data.head()
```

```python
from sklearn.preprocessing import LabelEncoder
label_columns = [
    'Patient_Gender',
    'Admission_Type',
    'Service_Type'
]



le = LabelEncoder()

for col in label_columns:
    data[col] = le.fit_transform(data[col])
```

```python
data.head()
```

```python
onehot_columns = [
    'Provider_Type',
    'Provider_Specialty',
    'Discharge_Type'
]
data = pd.get_dummies(
    data,
    columns=onehot_columns,
    drop_first=True
)
```

```python
data.head()
```

```python
high_cardinality_columns = [
    'Patient_City',
    'Provider_City',
    'Diagnosis_Code',
    'Procedure_Code'
]

for col in high_cardinality_columns:

    freq = data[col].value_counts()

    data[col] = data[col].map(freq)
```

```python
data.head()
```

```python
data['Claim_Year'] = data['Claim_Date'].dt.year
data['Claim_Month'] = data['Claim_Date'].dt.month

data['Claim_Delay_Days'] = (
    data['Claim_Date'] - data['Service_Date']
).dt.days
```

```python
data.drop(columns=[
    'Claim_Date',
    'Service_Date',
    'Policy_Expiration_Date'
], inplace=True)
```

```python
data.head(10)
```

```python
bool_cols = data.select_dtypes(include='bool').columns

data[bool_cols] = data[bool_cols].astype(int)
```

```python
data.head(5)
```

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()

data['Patient_State'] = le.fit_transform(data['Patient_State'])

data['Provider_State'] = le.fit_transform(data['Provider_State'])
```
