```python
drop_columns = [
    'Patient_ID',
    'Policy_Number',
    'Claim_ID',
    'Hospital_ID'
]

data.drop(columns=drop_columns, inplace=True)
```

```python
data
```

```python
data['Procedure_Code'] = data['Procedure_Code'].astype(str)
```

```python
numerical_columns = data.select_dtypes(include=['int64', 'float64']).columns

print(numerical_columns)
```

```python
for col in numerical_columns:

    Q1 = data[col].quantile(0.25)
    Q3 = data[col].quantile(0.75)

    IQR = Q3 - Q1

    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    outliers = data[
        (data[col] < lower_bound) |
        (data[col] > upper_bound)
    ]

    print(f"{col}")
    print(f"Number of outliers: {len(outliers)}")
    print("-" * 50)
```

```python
Q1 = data['Length_of_Stay_Days'].quantile(0.25)
Q3 = data['Length_of_Stay_Days'].quantile(0.75)

IQR = Q3 - Q1

lower = Q1 - 1.5 * IQR
upper = Q3 + 1.5 * IQR

data['Length_of_Stay_Days'] = data['Length_of_Stay_Days'].clip(lower, upper)
```
